```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load data from an Excel file
data = pd.read_excel("/content/Road Accident Data.xlsx")

# Display the first few rows of data
print(data.head())

# Check the data summary
print(data.info())

# Create a scatter plot to visualize the relationship between two variables
plt.scatter(data['Number_of_Persons in Accidents'], data['Number_of_Facalities'])
plt.title('Accidents vs Fatalities')
plt.xlabel('Number of Persons in Accidents')
plt.ylabel('Fatalities')
plt.show()

# Create a linear regression model
model = LinearRegression()

# Prepare the data for modeling
X = data[['Number_of_Persons in Accidents']]
y = data['Number_of_Facalities']

# Fit the linear regression model
model.fit(X, y)

# Create a scatter plot
plt.scatter(data['Number_of_Persons in Accidents'], data['Number_of_Facalities'], label='Data')
plt.title('Accidents vs Fatalities')
plt.xlabel('Number of Persons in Accidents')
plt.ylabel('Fatalities')
# Fit the linear regression model
model.fit(X, y)

# Create a scatter plot
plt.scatter(data['Number_of_Persons in Accidents'], data['Number_of_Facalities'], label='Data')
plt.title('Accidents vs Fatalities')
plt.xlabel('Number of Persons in Accidents')
plt.ylabel('Fatalities')

# Plot the linear regression line
plt.plot(X, model.predict(X), color='red', label='Linear Regression')
plt.legend()
plt.show()

# Get the coefficients
intercept = model.intercept_
slope = model.coef_[0]

# Print the regression equation
print(f"Regression Equation: Fatalities = {slope:.2f} * Number of Persons in Accidents + {intercept:.2f}")

# Create a scatter plot
plt.scatter(data['Number_of_Persons in Accidents'], data['Number_of_Facalities'], label='Data')
plt.title('Accidents vs Fatalities')
plt.xlabel('Number of Persons in Accidents')
plt.ylabel('Fatalities')

# Plot the linear regression line with the equation
plt.plot(X, model.predict(X), color='red', label=f'Linear Regression: Fatalities = {slope:.2f} * Number_of_Persons in Accidents + {intercept:.2f}')
plt.legend()
plt.show()

# Make a prediction
persons_involved = 75
predicted_fatalities = model.predict([[persons_involved]])
print(f"Predicted Fatalities for {persons_involved} persons involved in accidents: {predicted_fatalities[0]}")
```
